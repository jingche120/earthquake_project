```python
import pandas as pd

df = pd.read_csv("./source/all_metadata.csv")

# 一樣先去除重複的測站資料
df_unique = df.drop_duplicates(subset=['station_code'])

# 建立空字典，然後逐列塞入資料
station_coords_dict = {
    row['station_code']: [round(row['station_latitude_deg'],4), round(row['station_longitude_deg'],4)]
    for index, row in df_unique.iterrows()
}

print(station_coords_dict)

```

```python
import json
with open('../目前模型/source/our_parameters.json', 'w', encoding='utf-8') as f:
    json.dump(station_coords_dict, f, ensure_ascii=False, indent=4)
```

```python
with open('../目前模型/source/[全台]測站資料與位置.txt','w',encoding='utf-8') as f:
    for station,coords in station_coords_dict.items():
        lat = coords[0]
        lon = coords[1]
        line_text = f"Station: {station} Location: {lat:.2f} {lon:.2f}"
        f.write(line_text+'\n')
```

```python
# 假設這是你前面的字典資料
# station_coords_dict = {'ZUZH': [25.1621, 121.5447], 'EGFH': [23.7200, 121.5400]}

# 1. 打開一個名為 data.txt 的文字檔準備寫入 ('w' 模式)
with open('data.txt', 'w', encoding='utf-8') as f:
    
    # 2. 用 for 迴圈把字典裡的每一筆資料抽出來
    # station 會拿到 'ZUZH'，coords 會拿到 [25.1621, 121.5447]
    for station, coords in station_coords_dict.items():
        
        # 將串列中的經緯度分別取出來
        lat = coords[0]
        lon = coords[1]
        
        # 3. 組合你要的文字格式
        # 這裡的 :.4f 可以確保就算數字結尾是 0，也一定會印出小數點後四位
        line_text = f"Station: {station} Location: {lat:.4f} {lon:.4f}"
        
        # 4. 寫入檔案，並加上 \n 代表「換行」
        f.write(line_text + '\n')

print("資料已經成功寫入 data.txt 囉！")
```

```python
import pandas as pd

# 假設你的 DataFrame 叫做 df
data = pd.read_csv("./source/all_metadata.csv")
df = pd.DataFrame(data)
# 1. 找出每個 event_id 中 pga 最大的那「整筆」資料
# groupby('event_id')['pga'].idxmax() 會找出每個群組中 pga 最大的那個 index (索引值)
# 然後我們用 .loc[] 依照這些 index 把完整的 row 提取出來
max_pga_df = df.loc[df.groupby('source_event_id')['pga'].idxmax()]
print(f"max_pga_df{max_pga_df}")
print("-"*20)
# 2. 設定經緯度的篩選條件 (直接在程式裡做度分轉換，確保精準度)
lat_min = 23 + 15 / 60   # 23.25
lat_max = 24 + 25 / 60   # 24.4166...
lon_min = 121 + 5 / 60   # 121.0833...
lon_max = 121 + 45 / 60  # 121.75

# 3. 建立布林條件 (Boolean Indexing)
condition = (
    (max_pga_df['station_latitude_deg'] >= lat_min) &
    (max_pga_df['station_latitude_deg'] <= lat_max) &
    (max_pga_df['station_longitude_deg'] >= lon_min) &
    (max_pga_df['station_longitude_deg'] <= lon_max)
)

# 4. 篩選資料，只取出 'source_event_id' 欄位，並存成 List
target_event_ids = max_pga_df[condition]['source_event_id'].tolist()

# 印出結果看看
print(f"總共找到 {len(target_event_ids)} 筆符合條件的地震事件。")
print(target_event_ids)
```

```python
import pandas as pd

# 假設 target_event_ids 是你剛剛找出來的清單
# target_event_ids = [12010403340, 12010403341, ...]

# 1. 讀取原始的龐大 CSV 檔案
# 請將 'original_data.csv' 換成你實際的檔案路徑
df = pd.read_csv('../目前模型/source/all_metadata.csv', low_memory=False)

# 💡【重要防呆機制】：確保兩邊的資料型態一致！
# 有時候 CSV 讀進來的 ID 會變成字串，而你的 target_event_ids 可能是整數。
# 為了避免「123」和 123 比對失敗，我們統一把它們都轉成字串。
df['source_event_id'] = df['source_event_id'].astype(str).str.strip()
target_event_ids_str = [str(eid).strip() for eid in target_event_ids]

# 2. 核心動作：使用 .isin() 進行名單比對篩選
filtered_df = df[df['source_event_id'].isin(target_event_ids_str)]

# 印出篩選前後的資料筆數，讓自己安心
print(f"[*] 原始資料共 {len(df)} 筆")
print(f"[+] 篩選後剩下 {len(filtered_df)} 筆符合條件的資料")

# 3. 存成全新的 CSV 檔案
# index=False 的作用是不要把最左邊的 0, 1, 2, 3 索引號碼也存進 CSV 裡
output_filename = './source/filtered_by_Hualien_events.csv'
filtered_df.to_csv(output_filename, index=False, encoding='utf-8')

print(f"[完成] 專屬資料集已成功存至 {output_filename}")
```
