```python
import pandas as pd
import numpy as np

def add_intensity_and_overwrite(csv_path):
    """
    讀取 CSV 檔案，新增震度 (intensity) 欄位後，直接複寫原檔案。
    """
    print(f"[*] 正在讀取並處理資料: {csv_path}")
    df = pd.read_csv(csv_path)

    if 'pga' not in df.columns:
        raise ValueError(f"[!] 錯誤：{csv_path} 中找不到 'pga' 欄位！")

    # 定義 PGA 分級邊界與標籤
    bins = [-np.inf, 0.8, 2.5, 8.0, 25.0, 80.0, 140.0, 250.0, 440.0, 800.0, np.inf]
    labels = ['0 級', '1 級', '2 級', '3 級', '4 級', '5 弱', '5 強', '6 弱', '6 強', '7 級']

    # 新增 intensity 欄位
    df['intensity'] = pd.cut(df['pga'], bins=bins, labels=labels, right=False)

    # 🟢 直接使用原本的 csv_path 進行儲存，達成複寫覆蓋
    df.to_csv(csv_path, index=False)
    
    print(f"[+] 成功！已將震度欄位寫入並覆蓋原檔案：{csv_path}")
    
    # 印出一下分佈狀況讓你確認
    print("\n[📊] 資料庫目前的震度分佈：")
    print(df['intensity'].value_counts().sort_index())

# ==========================================
# 執行區塊
# ==========================================
if __name__ == "__main__":
    # 替換成你要複寫的 CSV 檔案路徑
    target_csv = '/Users/lin/碩班/地震模型/目前模型/source/filtered_hualien_taipei_events.csv'  
    
    add_intensity_and_overwrite(target_csv)
```
