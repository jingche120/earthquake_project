```python
import json
# 這是你提供的測站字典 (我這裡省略部分，你實際執行時貼上完整的)
input_json_path = r'C:\Users\user\Desktop\L_Earthquake_Model\earthquake_project\source\all_stastion_info.json'
with open(input_json_path, 'r', encoding='utf-8') as f:
    # 🌟 關鍵指令：json.load()
    loaded_data = json.load(f)

# 1. 設定花蓮的經緯度範圍 (沿用你之前的設定)
lat_min = 23 + 15 / 60   # 23.25
lat_max = 24 + 25 / 60   # 24.4166...
lon_min = 121 + 5 / 60   # 121.0833...
lon_max = 121 + 45 / 60  # 121.75

# 2. 準備兩個空清單來裝測站名稱
hualien_stations = []
other_stations = []

# 3. 掃描所有測站進行分類
for station_name, coords in loaded_data.items():
    lat, lon = coords[0], coords[1]
    
    # 判斷是否落在花蓮範圍內
    if (lat_min <= lat <= lat_max) and (lon_min <= lon <= lon_max):
        hualien_stations.append(station_name)
    else:
        other_stations.append(station_name)

# 4. 把兩個清單接起來：花蓮站排前面，其他站排後面
all_stations_sorted = hualien_stations + other_stations

# 印出結果檢查
print(f"總測站數: {len(all_stations_sorted)}")
print(f"花蓮測站數: {len(hualien_stations)} (已排在最前面 0 ~ {len(hualien_stations)-1} 的位置)")
print("-" * 30)
print("排序後的前 10 個測站 (應該都是花蓮):")
print(all_stations_sorted[:10])

# 5. 依照新的順序，建立一個全新的字典
sorted_station_data = {}
for station_name in all_stations_sorted:
    # 把原本 loaded_data 裡的經緯度資料，按照新順序塞進新字典
    sorted_station_data[station_name] = loaded_data[station_name]

# 6. 將重新排序好的字典，寫回成 JSON 檔案
# 建議先存成一個新的檔名，確認沒問題後再蓋掉原本的

with open(input_json_path, 'w', encoding='utf-8') as f:
    # indent=4 讓存出來的 JSON 檔案有漂亮的縮排，不會全部擠在同一行
    json.dump(sorted_station_data, f, indent=4, ensure_ascii=False)

print("-" * 30)
print(f"[+] 成功！已將重新排序的測站資料儲存至: {input_json_path}")



with open(r'C:\Users\user\Desktop\L_Earthquake_Model\earthquake_project\source\hualien_stations_info.json', 'w', encoding='utf-8') as f:
    # indent=4 讓存出來的 JSON 檔案有漂亮的縮排，不會全部擠在同一行
    json.dump(hualien_stations, f, indent=4, ensure_ascii=False)

print("-" * 30)
print(f"[+] 成功！已將重新排序的測站資料儲存至: hualien_stations_info.json")
```
